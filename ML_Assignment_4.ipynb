<b>3. Distinguish:<br>

1. Numeric vs. categorical attributes<br>

2. Feature selection vs. dimensionality reduction

Numeric vs. categorical attributes.<br>
Numeric variables are numbers which measure something both discrete and continuous like age, number of children. Categorical variable are variables having categories like sex, educational level

Feature selection vs. dimensionality reduction<br>
Feature selection is a process of selecting important features that can contribute to a model performance. Feature selection is done using the strength of association of an input feature with the target feature. Feature selection helps reduce overfitting by eliminating irrelevant features.

<br><br>Dimensionalty reduction is used to combine correlated features into one and reduce the number of features in a data set. Dimensionality reduction helps in quicker model training and reduce the size of training dataset.
